```python
import pandas as pd
import numpy as np
```

```python
pm_bkk = pd.read_csv("../../data/a1/bkk_pm.csv", parse_dates=["date"])
pm_bkk
```

```python
w_bkk = pd.read_csv('../../data/weather/bangkok_weather.csv', parse_dates=["date"])
w_bkk
```

```python
pm_bkk.set_index('date', inplace=True)
w_bkk.set_index('date', inplace=True)

w_bkk_monthly = w_bkk.resample('M').mean()
w_bkk_monthly
```

```python
w_bkk_monthly = w_bkk_monthly.drop(columns=['precipitation'])
```

```python
w_bkk_monthly.info()
```

```python
w_bkk_monthly
```

```python
# print(w_bkk_monthly[w_bkk_monthly.isnull().any(axis=1)])
```

```python
pm_bkk = pm_bkk.merge(w_bkk_monthly, left_index=True, right_index=True)
pm_bkk.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(pm_bkk.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between PM2.5 and Weather')
plt.show()
```

I set the threshould to use as features

```python
correlation_matrix = pm_bkk.corr(numeric_only=True)

target_corr = correlation_matrix['avg'].drop('avg')

selected_features = target_corr[abs(target_corr) > 0.3].index.tolist()

print(target_corr[abs(target_corr) > 0.3])
```

```python
selected_features = [
    'temperature_min',
    'dew_point_max',
    'dew_point_avg',
    'dew_point_min',
    'humidity_max',
    'humidity_avg',
    'humidity_min',
    'wind_speed_max',
    'wind_speed_avg',
    'pressure_max'
]

import matplotlib.pyplot as plt

pm_bkk.index = pd.to_datetime(pm_bkk.index)
pm_bkk[selected_features].plot(
    subplots=True,
    figsize=(12, 10),
    layout=(5, 2),
    title='Selected Features Correlation with PM2.5'
)

plt.tight_layout()
plt.show()
```

ok as I see. for dew_point and humidity all are selected, so I will use the avg of its. on the other hand, for wind speed and pressure which are not all selected. I will choose the windspeed max for significantly higher correlation temperature min and pressure max as features

```python
from pandas.plotting import autocorrelation_plot

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(14, 12))
axes = axes.flatten()

# Plot autocorrelation for each feature
for i, feature in enumerate(selected_features):
    autocorrelation_plot(pm_bkk[feature].dropna(), ax=axes[i])
    axes[i].set_title(f'Autocorrelation: {feature}')

# Hide unused subplot if any
for j in range(len(selected_features), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
features = [
    'temperature_min',
    'dew_point_avg',
    'humidity_avg',
    'wind_speed_max',
    'pressure_max'
]
```

```python
pm_bkk.index = pd.to_datetime(pm_bkk.index)
pm_bkk[features].plot(
    subplots=True,
    figsize=(12, 10),
    layout=(3, 2),
    title='Features Correlation with PM2.5'
)

plt.tight_layout()
plt.show()
```

```python
from sklearn.impute import KNNImputer

def impute_feature_knn(df, features, n_neighbors=5):
    imputer = KNNImputer(n_neighbors=n_neighbors)
    df[features] = imputer.fit_transform(df[features])
    return df

pm_bkk_f = impute_feature_knn(pm_bkk, features)
```

```python
pm_bkk_f.index = pd.to_datetime(pm_bkk_f.index)
pm_bkk_f[features].plot(
    subplots=True,
    figsize=(12, 10),
    layout=(3, 2),
    title='Features Correlation with PM2.5'
)

plt.tight_layout()
plt.show()
```
