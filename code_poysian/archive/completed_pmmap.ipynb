```python
import numpy as np
import pandas as pd
```

```python
df_pmmap = pd.read_csv("../../data/a1/PM_map.csv")
```

```python
dfs_pm = {}

for i in range(2011, 2025):
    dfs_pm[i] = pd.read_excel(f"../../data/pm2-5/Raw_Data/PM2.5/PM2.5({i}).xlsx", sheet_name="PM2.5")
```

```python
dfs_pm[2024]
```

```python
mapper = df_pmmap[['รหัสสถานี', 'ชื่อสถานี']]
mapper
```

```python
new_column = []

mapone = mapper["รหัสสถานี"]
maptwo = mapper["ชื่อสถานี"]
columns_list = dfs_pm[2024].columns.tolist()

for col in columns_list[1:]:
    if col in mapone.values:
        index = mapone[mapone == col].index[0]
        new_column.append(f"{col} | {maptwo[index]}")
    else:
        new_column.append(col)

# Update the columns of the DataFrame
dfs_pm[2024].columns = [columns_list[0]] + new_column
```

```python
#now do all files
for i in range(2011, 2025):
    mapper = df_pmmap[['รหัสสถานี', 'ชื่อสถานี']]
    new_column = []

    mapone = mapper["รหัสสถานี"]
    maptwo = mapper["ชื่อสถานี"]
    columns_list = dfs_pm[i].columns.tolist()

    for col in columns_list[1:]:
        if col in mapone.values:
            index = mapone[mapone == col].index[0]
            new_column.append(f"{col} | {maptwo[index]}")
        else:
            new_column.append(col)

    # Update the columns of the DataFrame
    dfs_pm[i].columns = [columns_list[0]] + new_column
```

```python
dfs_pm[2011]
```

```python
# Combine all years into a single DataFrame
df_all_years = pd.concat(dfs_pm.values(), keys=dfs_pm.keys(), names=['Year', 'Row'])

# Reset the index to make it easier to work with
df_all_years.reset_index(level=0, inplace=True)

df_all_years
```
