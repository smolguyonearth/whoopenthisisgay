```python
#plot graph of all pm2.5 in bangkok
#indicated by [col for col in pm_map.columns if col.split(" ")[-1] == "กทม."]
#date and time are in the same column 2024-12-31 00:00:00
import matplotlib.pyplot as plt
import seaborn as sns

def plot_pm2_5(c):
    pm_map_local = pm_map.copy()
    pm_map_local["date"] = pd.to_datetime(pm_map_local["Date"])
    pm_map_local.set_index("date", inplace=True)
    
    # Filter Bangkok columns
    pm_map_local = pm_map_local.loc[:, [col for col in pm_map_local.columns if col.split(" ")[-1] == c]]
    
    # Convert to numeric
    pm_map_local = pm_map_local.apply(pd.to_numeric, errors='coerce')
    
    # Resample Daily
    pm_map_local = pm_map_local.resample("D").mean()
    
    pm_map_local.columns = [col.split(" ")[0] for col in pm_map_local.columns]
    
    pm_map_local.plot(figsize=(20, 10))
    plt.title("PM2.5 in Bangkok")
    plt.xlabel("Date")
    plt.ylabel("PM2.5")
    plt.show()
    print(pm_map_local.tail())

plot_pm2_5("กทม.")
```

```python
df_54T = pm_map.loc[:, [col for col in pm_map.columns if col.split(" ")[0] == "54T"]]
# add column date back
df_54T["date"] = pm_map["Date"]
```

```python
from sklearn.impute import KNNImputer

# Exclude the 'date' column for imputation
df_54T_numeric = df_54T.drop(columns=["date"])

# Create a KNNImputer object
imputer = KNNImputer(n_neighbors=5)

# Fit and transform the numeric data
df_54T_imputed_numeric = imputer.fit_transform(df_54T_numeric)

# Convert the result back to a DataFrame
df_54T_imputed = pd.DataFrame(df_54T_imputed_numeric, columns=df_54T_numeric.columns)

# Add the 'date' column back
df_54T_imputed["date"] = pd.to_datetime(df_54T["date"])

# Plot the imputed data
pm_map_local = df_54T_imputed.copy()
pm_map_local.set_index("date", inplace=True)

# Convert to numeric
pm_map_local = pm_map_local.apply(pd.to_numeric, errors='coerce')

# Resample Daily
pm_map_local = pm_map_local.resample("D").mean()

pm_map_local.plot(figsize=(20, 10))
plt.title("PM2.5 in Bangkok")
plt.xlabel("Date")
plt.ylabel("PM2.5")
plt.show()
print(pm_map_local.tail())
```

```python
#for all stations in bangkok use avg of all stations
def plot_pm2_5_all_stations():
    pm_map_local = pm_map.copy()
    pm_map_local["date"] = pd.to_datetime(pm_map_local["Date"])
    pm_map_local.set_index("date", inplace=True)
    
    # Filter Bangkok columns
    pm_map_local = pm_map_local.loc[:, [col for col in pm_map_local.columns if col.split(" ")[-1] == "กทม."]]
    
    # Convert to numeric
    pm_map_local = pm_map_local.apply(pd.to_numeric, errors='coerce')
    
    # Resample Daily
    pm_map_local = pm_map_local.resample("D").mean()
    
    # Calculate the average across all stations
    pm_map_local["avg"] = pm_map_local.mean(axis=1)
    
    pm_map_local["avg"].plot(figsize=(20, 10))
    plt.title("PM2.5 in Bangkok (All Stations Average)")
    plt.xlabel("Date")
    plt.ylabel("PM2.5")
    plt.show()
    print(pm_map_local.tail())


plot_pm2_5_all_stations()
```

```python
def plot_pm2_5_all_stations():
    pm_map_local = pm_map.copy()
    pm_map_local["date"] = pd.to_datetime(pm_map_local["Date"])
    pm_map_local.set_index("date", inplace=True)

    # Filter Bangkok columns
    pm_map_local = pm_map_local.loc[:, [col for col in pm_map_local.columns if col.split(" ")[-1] == "กทม."]]

    # Convert to numeric
    pm_map_local = pm_map_local.apply(pd.to_numeric, errors='coerce')

    # Resample Daily
    pm_map_local = pm_map_local.resample("M").mean()

    # Calculate the average across all stations
    pm_map_local["avg"] = pm_map_local.mean(axis=1)

    # Plot
    pm_map_local["avg"].plot(figsize=(20, 10))
    plt.title("PM2.5 in Bangkok (All Stations Average)")
    plt.xlabel("Date")
    plt.ylabel("PM2.5")
    plt.show()

    print(pm_map_local.tail())
    
    return pm_map_local

pm_map_bangkok = plot_pm2_5_all_stations()
series = pm_map_bangkok["avg"].dropna()

# Save the series to a CSV file
series.to_csv("../../data/a1/pm_bangkok_avg.csv", index=True)
```
