```python
from sklearn.impute import KNNImputer
import pandas as pd

# Only select numeric columns
df_numeric = df_chb.select_dtypes(include=['number'])

# Drop the 'precipitation' column as it contains only NaN values
df_numeric_without_precipitation = df_numeric.drop(columns=['precipitation'])

# Impute only numeric part
imputer = KNNImputer(n_neighbors=5)
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df_numeric_without_precipitation), columns=df_numeric_without_precipitation.columns)

# Add the 'precipitation' column back with NaN values
df_numeric_imputed['precipitation'] = df_numeric['precipitation'].values

# Now if you want to combine back with the original non-numeric columns:
df_non_numeric = df_chb.select_dtypes(exclude=['number']).reset_index(drop=True)
df_imputed = pd.concat([df_numeric_imputed, df_non_numeric], axis=1)

# Example access
print(df_imputed.loc[31])
```

```python
df_imputed['date'] = pd.to_datetime(df_imputed['date'], dayfirst=True)
df_imputed = df_imputed.sort_values('date')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import numpy as np

# Step 2: Select the temperature series you want to predict
series = df_imputed['temperature_avg']

# Step 3: Train ARIMA model
model = ARIMA(series, order=(1, 1, 1))  # (p, d, q) can be adjusted
model_fit = model.fit()

# Step 4: Forecast the next 30 days
forecast_steps = 30
forecast = model_fit.forecast(steps=forecast_steps)

# Step 5: Generate forecast dates
forecast_dates = pd.date_range(series.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')

# Step 6: Assume you have actual observed data for the next 30 days (if available)
# For demonstration, let's assume you have actual data in 'actual_data' list
# Replace this with your actual observed data if available
actual = df_imputed['temperature_avg'][-30:]  # Example placeholder (use actual values if available)

# Step 7: Compute accuracy metrics
# 1. Mean Absolute Error (MAE)
mae = mean_absolute_error(actual, forecast)

# 2. Mean Squared Error (MSE)
mse = mean_squared_error(actual, forecast)

# 3. Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# 4. Mean Absolute Percentage Error (MAPE)
mape = mean_absolute_percentage_error(actual, forecast)

# 5. R-squared (RÂ²)
ss_residual = np.sum((actual - forecast) ** 2)
ss_total = np.sum((actual - np.mean(actual)) ** 2)
r2 = 1 - (ss_residual / ss_total)

# Print results
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}")
print(f"R-squared (RÂ²): {r2:.2f}")

# Step 8: Plotting the observed and forecasted values
plt.figure(figsize=(14, 7))
plt.plot(series, label='Observed', color='blue')
plt.plot(forecast_dates, forecast, label='Forecast (Next 30 Days)', color='red', linestyle='--')
plt.title('Temperature Forecast with ARIMA')
plt.xlabel('Date')
plt.ylabel('Temperature (F)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import itertools
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Define a function to evaluate ARIMA models with different parameter combinations
def evaluate_arima_models(series, p_values, d_values, q_values):
    best_score, best_cfg = float('inf'), None
    best_model = None
    for p, d, q in itertools.product(p_values, d_values, q_values):
        try:
            model = ARIMA(series, order=(p, d, q))
            model_fit = model.fit()
            predictions = model_fit.forecast(steps=30)  # predict next 30 days
            error = mean_squared_error(series[-30:], predictions)  # Use MSE to evaluate performance
            if error < best_score:
                best_score, best_cfg = error, (p, d, q)
                best_model = model_fit
        except Exception as e:
            continue
    return best_score, best_cfg, best_model

# Define parameter grid for tuning
p_values = range(0, 5)  # AR terms (p)
d_values = range(0, 2)  # Differencing terms (d)
q_values = range(0, 5)  # MA terms (q)

# Evaluate the ARIMA models
best_score, best_cfg, best_model = evaluate_arima_models(series, p_values, d_values, q_values)
print(f"Best ARIMA Model: ARIMA{best_cfg} with MSE={best_score}")

# Use the best model to forecast
forecast = best_model.forecast(steps=30)
```

```python
# Print the AIC and BIC for the selected model
print(f"AIC: {best_model.aic}")
print(f"BIC: {best_model.bic}")
```

```python
# Forecast the next 30 days using the best model
forecast = best_model.forecast(steps=30)

# Print the forecasted values
print("\nForecast for the next 30 days:")
print(forecast)

# Plot the results
plt.figure(figsize=(14,7))
plt.plot(series, label='Observed')
plt.plot(pd.date_range(series.index[-1], periods=30, freq='D'), forecast, label='Forecast (Next 30 Days)', color='red', linestyle='--')
plt.title('Temperature Forecast with Tuning ARIMA')
plt.xlabel('Date')
plt.ylabel('Temperature (F)')
plt.legend()
plt.grid(True)
plt.show()
```
