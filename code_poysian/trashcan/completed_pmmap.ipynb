```python
for i in range(2024, 2010, -1):
    mapper = df_pmmap.set_index('รหัสสถานี')['ชื่อสถานี'].to_dict()
    
    # Rename columns
    new_columns = []
    for col in dfs_pm[i].columns:
        print(f"Processing column: {col}")
        if col in mapper:
            new_columns.append(f"{col} | {mapper[col]}")
            print(f"Renamed {col} to {col} | {mapper[col]}")
        else:
            new_columns.append(col)
    
    dfs_pm[i].columns = new_columns
    
    # Remove last 2 rows
    dfs_pm[i] = dfs_pm[i].iloc[:-2]
```

```python
# Combine all years into a single DataFrame
df_all_years = pd.concat(dfs_pm.values(), keys=dfs_pm.keys(), names=['Year', 'Row'])

# Reset the index to make it easier to work with
df_all_years.reset_index(level=0, inplace=True)

# df_all_years
```
